Question 7: Write a Python program to find the longest word in a string.
sample input :
var = 'python is Easy Language'
sample output :
result : Language

```python
var = 'python is Easy Language'
a=var.split()
heighest=a[0]
for word in a:
  if len(word) >len(heighest):
    heighest=word

print(heighest)
```
