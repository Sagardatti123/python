### Question 1: Write a function that takes a dictionary and a key as input and returns the corresponding value.

- `Sample Input: my_dict = {'a': 1, 'b': 2, 'c': 3}, key = 'b'`
- `Sample Output: 2`

```python
# Please write The solution Here for above question 1:
n=int(input("enter no.of key value pairs:"))
my_dict={}
for i in range(n):
  key=input("enter a key:")
  value=int(input("enter a value:"))
  my_dict[key]=value
print(my_dict)
user=input("enter a key:")
if user in my_dict:
  print(my_dict[user])
```

### Question 9 : Write a function that sorts a dictionary by its values in ascending order.
- `Sample Input: my_dict = {'a': 5, 'b': 2, 'c': 7}`

```python
# Please write The solution Here for above question 9:
### Question 9 : Write a function that sorts a dictionary by its values in ascending order.

my_dict = {'a': 5, 'b': 2, 'c': 7}
my_dict1={}
a=sorted(list(my_dict.values()))  #257
for i in a:
  for key,value in my_dict.items():
    if value==i:
      my_dict1[key]=value
print(my_dict1)



```

### Question 10 : Write a function that returns the key in a dictionary with the maximum value.
- `Sample Input: my_dict = {'a': 5, 'b': 2, 'c': 7}`
- `Sample Output: 'c'`

```python
my_dict = {'a': 5, 'b': 2, 'c': 7}

for key,value in my_dict.items():
  if value==max(list(my_dict.values())):
    print(key)
```

```python
my_dict = {
    'cat': 10,
    'apple': 20,
    'ball': 30,
    'dog': 40
}

a=list(my_dict.keys())
for i in range(len(a)):
  min_index=i
  for j in range(i+1,len(a)):
    if a[j]<a[min_index]:
      min_index=j
  temp=a[i]
  a[i]=a[min_index]
  a[min_index]=temp

for key in a:
  print(key)


```
